# Hierarchical Clustering

```python
# import Libraries
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

## Import dataset

```python
# Import the dataset from csv file
df = pd.read_csv('data/cluster2.csv')
print("Dimensions of the data = {}".format(df.shape))
df[:5]
```

## Preprocessing

```python
# convert df to array
X = df.values

# Set the size of the figure
plt.figure(figsize=(5,5))

# color-map
cmap = "tab10" 

# Plot the data points
plt.scatter(X[:,0], X[:,1],s=4, cmap = cmap)

plt.show() # print the figure
```

## Clustering model

```python
for i,linkage in enumerate(("single","complete","average","ward")) :
    clustering = AgglomerativeClustering(
        linkage = linkage,
        n_clusters = 4
    )
    y_pred = clustering.fit_predict(X)
    plt.figure(i+1,figsize = (5,5))
    plt.title(f"n_clusters : {4}, linkage : {linkage}")
    plt.scatter(X[:,0],X[:,1],c = y_pred, s = 4, cmap = cmap)
plt.show()

### Clustering parameters
## n_clusters : the number of clusteres to find
## linkage : the link type (single, complete, average, ward)
## affinity : the distance metric (default : euclidean)
```
